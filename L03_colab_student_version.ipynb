# Task 1: Creating a Simple MLP Class

### What you are learning
In this task, you are learning how to define a neural network by **subclassing `nn.Module`**, which is the standard PyTorch pattern for custom models. You will create layers in `__init__()` and define how data flows through them in `forward()`. This separation is important because the model architecture (layers) should be created once, while the forward pass runs every time you pass data through the model.

### Steps (from the lab)
1. Import the PyTorch module.
2. Define a simple multi-layer perceptron (MLP) class.
3. Define the first linear layer.
4. Define activation function.
5. Define the second linear layer.
6. Define the forward pass logic.

> **Note:** The lab PDF has a capitalization typo (`Self.relu`). In Python, it should be `self.relu`.


```python
# 1. Import the PyTorch module.
import torch
import torch.nn as nn

# 2. Define a simple multi-layer perceptron (MLP) class.
class SimpleMLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleMLP, self).__init__()

        # 3. Define the first linear layer.
        self.fc1 = nn.Linear(input_size, hidden_size)

        # 4. Define activation function.
        self.relu = nn.ReLU()

        # 5. Define the second linear layer.
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # 6. Forward pass logic.
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

### Task 1 explanation (learning takeaways)
- `nn.Module` is the base class for custom PyTorch models.
- `nn.Linear` creates fully connected layers.
- `nn.ReLU()` adds nonlinearity so the model can learn more than a simple linear mapping.
- `forward()` defines the exact order of operations for the input tensor.


# Task 2: Instantiating the Model and Inspecting Parameters

### What you are learning
This task teaches you how to **create an instance of your model** and inspect what PyTorch has built. Printing the model helps you verify the architecture, and using `named_parameters()` helps you see each trainable tensor (weights and biases), its shape, and whether gradients are enabled. This is a key debugging skill when building custom networks.

### Steps (from the lab)
1. Instantiate the model with example values for input, hidden, and output sizes.
2. Print the model and display parameter details to verify successful creation.
3. Review the output (the lab PDF shows an example screenshot of the printed model and parameter sizes on page 3).


```python
# 1. Instantiate the model with some values of the input layer,
#    hidden layer, and output layer.
model = SimpleMLP(input_size=10, hidden_size=5, output_size=1)

# 2. Print the model and verify parameters and successful creation.
print(model)
print("-" * 60)

for name, param in model.named_parameters():
    print(f"Layer: {name} | Size: {tuple(param.size())} | Requires_grad: {param.requires_grad}")
```

### Task 2 explanation (learning takeaways)
- `model = SimpleMLP(...)` creates a concrete network from your class definition.
- `print(model)` shows the module hierarchy (layers inside the model).
- `named_parameters()` lists each trainable parameter, including:
  - **weights** (matrix values used for transformations)
  - **biases** (offset values)
- The parameter shapes help confirm your layer dimensions are correct.

# Challenge: Add a Dropout Layer After ReLU

### What you are learning
This challenge extends your MLP by adding **dropout**, a regularization technique that randomly zeros some activations during training to reduce overfitting. You will modify the model, instantiate it, pass dummy input through it, and compare the printed architecture and output to the original model.

### Challenge prompt (from the lab)
Modify the MLP model to add a dropout layer after the ReLU activation. Then:

1. Instantiate the modified model.
2. Pass dummy input through it.
3. Print the model architecture and output.


```python
# Challenge solution cell (runnable example)

class SimpleMLPWithDropout(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, dropout_p=0.5):
        super(SimpleMLPWithDropout, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=dropout_p)  # Dropout added after ReLU
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# 1. Instantiate the modified model.
dropout_model = SimpleMLPWithDropout(input_size=10, hidden_size=5, output_size=1, dropout_p=0.5)

# 2. Pass dummy input through it.
dummy_input = torch.randn(2, 10)
dropout_model.train()  # Enable dropout behavior for demonstration
dropout_output = dropout_model(dummy_input)

# 3. Print the model architecture and output.
print(dropout_model)
print("-" * 60)
print(f"Model output with dropout (train mode):\n{dropout_output}")

# Optional comparison in eval mode (dropout disabled)
dropout_model.eval()
dropout_output_eval = dropout_model(dummy_input)
print("\nModel output with dropout model in eval mode (dropout disabled):")
print(dropout_output_eval)
```

### Challenge explanation (learning takeaways)
- `nn.Dropout` is often used during training to improve generalization.
- Dropout behaves differently in:
  - **train mode** (`model.train()`) → dropout active
  - **eval mode** (`model.eval()`) → dropout inactive
- This introduces the idea that model behavior can depend on training vs. inference mode.

**1. Role of `forward()`**  
`forward()` defines how input data flows through the model (the sequence of operations/layers used to compute the output).

**2. Why define layers in `__init__()`**  
Layers should be created once in `__init__()` so PyTorch can properly register their parameters for training, saving/loading, and optimization. Creating layers inside `forward()` would recreate them each pass and break training behavior.
