# Task 1: Defining Scalar Function and Backward Pass

### What you are learning
In this task, you are learning the **core idea of automatic differentiation** in PyTorch. A tensor with `requires_grad=True` tells PyTorch to track operations performed on it. When you call `.backward()` on a scalar output, PyTorch walks backward through the computation graph and calculates the derivative with respect to the input tensor. This is the foundation of training neural networks, where gradients guide parameter updates.

### Steps (from the lab)
1. Define input tensor with gradient tracking.
2. Define a simple scalar function.
3. Perform backward pass.
4. Inspect the gradient.
5. Review the output and explanation.


```python
# 1. Define input tensor with gradient tracking.
x = torch.tensor(2.0, requires_grad=True)

# 2. Define a simple scalar function.
y = x**2 + 3 * x

# 3. Perform backward pass.
y.backward()

# 4. Inspect the gradient.
print(f"Value of x: {x.item()}")
print(f"Value of y: {y.item()}")
print(f"Gradient dy/dx stored in x.grad: {x.grad.item()}")
```

### Task 1 explanation (learning takeaways)
- A scalar tensor `x` is created with `requires_grad=True` to track computation.
- A simple function `y = x² + 3x` is defined.
- `.backward()` triggers automatic differentiation.
- The `.grad` field of `x` stores the derivative of `y` with respect to `x`.


# Task 2: Inspecting `.grad` to Understand Its Meaning

### What you are learning
This task reinforces that **gradients depend on the input value**. Even though the function formula stays the same, changing `x` changes the slope (gradient) at that point. This helps connect `.grad` to the mathematical idea of a derivative as a **rate of change at a specific input**.

### Steps (from the lab)
1. Create a tensor with `requires_grad=True`.
2. Define the scalar function `y = x^2 + 3x`.
3. Perform backward pass.
4. Print the original value of `x` and the gradient.
5. Review what `.grad` means.


```python
# 1. Create a tensor with requires_grad=True to track gradients.
x = torch.tensor(5.0, requires_grad=True)

# 2. Define a simple scalar function: y = x^2 + 3x.
y = x**2 + 3 * x

# 3. Perform backward pass to compute dy/dx.
y.backward()

# 4. Print original value of x and gradient.
print(f"For x = {x.item()}, dy/dx = {x.grad.item()}")
```

### Task 2 explanation (learning takeaways)
- A new scalar input `x` is used.
- You compute `dy/dx` for a different input.
- `.grad` explicitly shows how Autograd computes gradients based on the formula.
- It reinforces that gradients vary depending on the input value.

# Task 5: Using `detach()` and `torch.no_grad()`

### What you are learning
This task teaches two key ways to **stop gradient tracking** in PyTorch. Use `.detach()` when you want a tensor value that is disconnected from the graph, and use `torch.no_grad()` when performing operations (like inference) where gradients are not needed. These techniques improve efficiency and prevent accidental graph growth.

### Steps (from the lab)
1. Create a tensor with gradient tracking enabled.
2. Define a simple function of `x`.
3. Detach `y` from the computation graph.
4. Print the detached tensor value.
5. Perform a computation inside a `torch.no_grad()` block.
6. Review the explanation.


```python
# 1. Create a tensor with gradient tracking enabled.
x = torch.tensor(3.0, requires_grad=True)

# 2. Define a simple function of x (y = x^2).
y = x**2

# 3. Detach y from the computation graph.
y_detached = y.detach()

# 4. Print the value of the detached tensor.
print(f"Detached tensor value: {y_detached.item()}")

# 5. Perform computation inside a no_grad() block.
with torch.no_grad():
    z = x**3
print(f"Result with no_grad(): {z.item()}")
```

### Task 5 explanation (learning takeaways)
- `.detach()` creates a tensor disconnected from the computation graph.
- It is useful for inference or saving intermediate values without tracking gradients.
- `torch.no_grad()` temporarily disables gradient tracking for operations inside its scope.
- It helps optimize performance during inference.


# Task 6: Demonstrating Gradient Accumulation vs. Clearing Gradients

### What you are learning
This task teaches an extremely important PyTorch behavior: **gradients accumulate by default**. If you call `.backward()` multiple times without clearing gradients, the new gradients are added to the old ones. In training loops, this is why you typically call `optimizer.zero_grad()` (or manually clear `.grad`) before the next backward pass.

### Steps (from the lab)
1. Create a tensor with gradient tracking enabled.
2. First backward pass.
3. Second backward pass without clearing.
4. Reset gradients manually.
5. Third backward pass after clearing.
6. Review the explanation.

> **Note:** The lab PDF shows `Y3` in one line; here we use `y3` consistently.


```python
# 1. Create a tensor with gradient tracking enabled.
x = torch.tensor(2.0, requires_grad=True)

# 2. First backward pass.
y1 = x**2
y1.backward()
print(f"First backward pass: x.grad = {x.grad.item()}")

# 3. Second backward pass without clearing.
y2 = x**3
y2.backward()
print(f"Second backward pass without clearing: x.grad = {x.grad.item()}")

# 4. Now reset gradients manually.
x.grad.zero_()

# 5. Third backward pass after clearing.
y3 = x**4
y3.backward()
print(f"Third backward pass after clearing: x.grad = {x.grad.item()}")
```

### Task 6 explanation (learning takeaways)
- PyTorch accumulates gradients by default.
- This is why clearing gradients (for example, `x.grad.zero_()` or `optimizer.zero_grad()`) is necessary.
- Without clearing, gradients become incorrect for the current step because they include previous steps.
- This teaches proper gradient management during training.
