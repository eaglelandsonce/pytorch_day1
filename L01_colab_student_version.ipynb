## Task 1: Creating tensors

### What you are learning
Here you are learning the basic building blocks of PyTorch: tensors. You will create tensors with specific values, random values, zeros, and ones. These patterns show up everywhere in deep learning—for example, input data, initialized weights, masks, and placeholders for intermediate results.

### What to watch for
Notice the **shape** (dimensions) of each tensor and how the values are generated.


```python
# Task 1 - Creating tensors

# 1. Create a 1D tensor of 5 elements
tensor_1d = torch.tensor([1, 2, 3, 4, 5])
print("1D Tensor:\n", tensor_1d)

# 2. Create a 2D tensor (3x3 matrix) with random values
tensor_2d = torch.randn(3, 3)
print("\n2D Tensor (random values):\n", tensor_2d)

# 3. Create a tensor of zeros
tensor_zeros = torch.zeros((2, 4))
print("\nZeros Tensor:\n", tensor_zeros)

# 4. Create a tensor of ones
tensor_ones = torch.ones((2, 4))
print("\nOnes Tensor:\n", tensor_ones)
```

## Task 2: Indexing, slicing, and reshaping tensors

### What you are learning
This section teaches you how to access parts of a tensor and change its shape without changing its underlying data values. These are foundational skills for preparing batches, selecting features, and reorganizing tensors for model input/output. If you can index and reshape confidently, you can solve many real data-prep problems.

### What to watch for
Pay attention to the difference between:
- **Indexing** (selecting one part)
- **Slicing** (selecting a range)
- **Reshaping** (changing the layout)


```python
# Task 2 - Indexing, slicing, reshaping

# 1. Create a 1D tensor [0, 1, ..., 11] and reshape to (3, 4)
x = torch.arange(12).reshape(3, 4)
print("Original Tensor:\n", x)

# 2. Indexing: Get the first row
print("\nFirst row:", x[0])

# 3. Slicing: Get the first 2 rows
print("\nFirst 2 rows:\n", x[:2])

# 4. Reshape to a 2x6 tensor
x_reshaped = x.view(2, 6)
print("\nReshaped Tensor (2,6):\n", x_reshaped)
```

## Task 3: Performing math operations

### What you are learning
In this task, you are learning the difference between **element-wise operations** and **matrix multiplication**. This is one of the most important distinctions in tensor programming. Element-wise math works position-by-position, while matrix multiplication combines rows and columns and is central to neural network computations.

### What to watch for
Compare the outputs of `a + b`, `a * b`, and `a @ b` and notice how the results differ.


```python
# Task 3 - Math operations

# 1. Create two 2x2 tensors
a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[5, 6], [7, 8]])

print("Tensor a:\n", a)
print("\nTensor b:\n", b)

# 2. Element-wise addition
add_result = a + b
print("\nElement-wise addition (a + b):\n", add_result)

# 3. Element-wise multiplication
mul_result = a * b
print("\nElement-wise multiplication (a * b):\n", mul_result)

# 4. Matrix multiplication
matmul_result = a @ b
print("\nMatrix multiplication (a @ b):\n", matmul_result)
```

## Task 4: Moving tensors to GPU

### What you are learning
This section introduces device management. PyTorch tensors live on a device (CPU or GPU), and operations usually require tensors to be on the same device. You are learning how to detect GPU availability and move tensors safely between devices, which is essential for scaling up to faster model training.

### What to watch for
Look at the `.device` shown for each tensor before and after moving it.


```python
# Task 4 - Move tensors between CPU and GPU

# Re-import torch if runtime was restarted after switching CPU/GPU
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Create a tensor on CPU
cpu_tensor = torch.tensor([1.0, 2.0, 3.0])
print("CPU tensor:", cpu_tensor, "| device:", cpu_tensor.device)

# Move tensor to selected device (GPU if available)
device_tensor = cpu_tensor.to(device)
print("Tensor on device:", device_tensor, "| device:", device_tensor.device)

# Optional: move back to CPU
back_to_cpu = device_tensor.to("cpu")
print("Moved back to CPU:", back_to_cpu, "| device:", back_to_cpu.device)
```

## Task 5: Broadcasting

### What you are learning
Broadcasting lets PyTorch perform operations on tensors with different shapes when the shapes are compatible. This saves time and avoids manually copying data. You are learning a powerful feature that is used often in normalization, bias addition, and many tensor transformations.

### What to watch for
Focus on how PyTorch expands shapes automatically during `x + y` without you explicitly reshaping both tensors.


```python
# Task 5 - Broadcasting

# 1. Broadcasting example
x = torch.ones(3, 1)
y = torch.ones(1, 4) * 2

print("x shape:", x.shape)
print("y shape:", y.shape)
print("x:\n", x)
print("y:\n", y)

# 2. Broadcast to (3,4) during addition
z = x + y
print("\nBroadcasted result z = x + y:\n", z)
print("z shape:", z.shape)
```

## Task 6: In-place operations

### What you are learning
In-place operations modify a tensor directly in memory instead of creating a new tensor. This can be efficient, but it also requires care because it changes the original tensor immediately. In training workflows, in-place changes can sometimes interfere with gradient tracking if used incorrectly.

### What to watch for
Notice how the tensor changes after `add_()` and that the original values are overwritten.


> **Student tip:** In-place operations often end with an underscore (for example, `add_()`). The underscore is your warning sign that the original tensor will be changed.


```python
# Task 6 - In-place operations

x = torch.tensor([1.0, 2.0, 3.0])
print("Original tensor:", x)

# In-place addition (modifies x directly)
x.add_(1)
print("After in-place addition:", x)
```

## Task 7: Seeding for reproducibility

### What you are learning
Randomness is useful in machine learning, but reproducibility is critical for debugging and comparing results. In this section, you are learning how `torch.manual_seed()` helps generate the same random values each time you rerun code. This makes experiments easier to verify and discuss.

### What to watch for
The two random tensors should match because the same seed is set before each one is created.


```python
# Task 7 - Reproducibility with manual seed

torch.manual_seed(42)
rand_tensor_1 = torch.randn(3, 3)
print("Random Tensor with Seed 42 (first run):\n", rand_tensor_1)

# Re-seed and regenerate to demonstrate reproducibility
torch.manual_seed(42)
rand_tensor_2 = torch.randn(3, 3)
print("\nRandom Tensor with Seed 42 (second run):\n", rand_tensor_2)

print("\nAre both random tensors identical?", torch.equal(rand_tensor_1, rand_tensor_2))
```

## Task 8: Saving and loading tensors

### What you are learning
This task shows how to save tensor data to disk and load it back later. This is a core workflow in PyTorch for checkpoints, cached preprocessing results, and saved intermediate outputs. Being able to persist tensors helps you avoid recomputing work and supports reproducible labs and experiments.

### What to watch for
The final equality check should return `True`, showing that the saved and loaded tensors match.


```python
# Task 8 - Saving and loading tensors

# 1. Create a tensor to save
x = torch.arange(10)
print("Original Tensor:\n", x)

# 2. Save the tensor to a file
torch.save(x, "tensor.pt")
print("\nTensor saved to tensor.pt")

# 3. Load back the tensor
x_loaded = torch.load("tensor.pt")
print("\nLoaded tensor:\n", x_loaded)

# 4. Verify equality
print("\nMatch:", torch.equal(x, x_loaded))
```
